## Random Forest Regression Model

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
kfold=TimeSeriesSplit(n_splits=10).split(train)
```

```python
params={'max_depth':[2,3,5],'n_estimators':[25,50,75,100]}
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor()
```

```python
gs = GridSearchCV(estimator= rfr, cv= kfold, param_grid= params)
#xval=train.drop(['DwellTime'],axis=1)
#yval=train['DwellTime']
gs.fit(xval, yval)
```

```python
gs.best_estimator_
```

```python
rfr.fit(xval,yval)
```

```python
gs.best_estimator_.fit(xval,yval)
```

# Model accuracy on the training set

```python
gs.best_estimator_.score(xval,yval)
```

# Model accuracy on the test test

```python
gs.best_estimator_.score(xtest,ytest)
```

```python
finalp=gs.best_estimator_.predict(xtest)
```

```python
finalp=rfr.predict(xtest)
```

```python
finalpr=finalp.round(0)
```

```python
finalp=pd.DataFrame(finalp)
finalp.rename(columns={0:'finalp'}, inplace=True)
finalpr=pd.DataFrame(finalpr)
finalpr.rename(columns={0:'finalpr'},inplace=True)
```

```python
finalp.reset_index(drop=True,inplace=True)
finalpr.reset_index(drop=True,inplace=True)
xtest.reset_index(drop=True,inplace=True)
ytest.reset_index(drop=True,inplace=True)
```

# Comparing actuals (Dwell Time) against model predictions (finalpr)

```python
table=pd.concat([xtest,ytest,finalp,finalpr],axis=1)
table
```

```python
features = xval.columns.tolist()
importances = gs.best_estimator_.feature_importances_
indices = np.argsort(importances)
```

# Importance given by the model to each input feature

```python
plt.title('Feature Importances')
indices1 = np.argsort(importances)
indices2=indices1[::-1]
plt.bar(range(len(indices)), importances[indices2], color='b', align='center')
plt.xticks(range(len(indices)), [features[i] for i in indices2],rotation=90)
#plt.set_xticklabels(rotation=90)
plt.ylabel('Relative Importance')
plt.show()
```

## XGBoost Regressor

```python
from xgboost import XGBRegressor
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
kfold=TimeSeriesSplit(n_splits=10).split(train)
```

```python
xgb = XGBRegressor(learning_rate=0.01, n_estimators=100,max_depth=5,nthread=4,seed=7)
```

```python
params={'max_depth':[2,3,5],'n_estimators':[25,50,75,100]}
```

```python
gs = GridSearchCV(estimator= xgb, cv= kfold, param_grid= params)
xval=train.drop(['DwellTime'],axis=1)
yval=train['DwellTime']
gs.fit(xval, yval)
```
