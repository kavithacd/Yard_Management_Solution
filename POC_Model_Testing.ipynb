## Container-Slot matching heuristic

```python
for i in range(length+1):
    for j in range(breadth+1):
        for k in range(height+1):
            ind=(i,j,k)
            for q in range(len(ordered['Container No'])):
                if(ordered['Container No'][q] in alloc.values()):
                    continue
                else:
                    if(alloc[ind]!=None):
                        continue
                    else:
                        alloc[ind]=ordered['Container No'][q]

alloc
```
